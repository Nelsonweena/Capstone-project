## Grid Adventure V1

### Overview

Grid Adventure V1 is, at its core, a grid‑world environment enriched with varied mechanics. The input is a MxN grid (positive integers M, N). Each tile can contain multiple objects (floor, agent, key, door, etc.), each with distinct behavior. The objective is collect any gems if present, before reaching the exit tile.

If you didn’t skip the getting started section at the top of this notebook, you should already be familiar with the game.

### Constraints (Project Scope)

To keep the capstone project tractable, we enforce the following constraints:

- There is exactly 1 agent in the grid, with starting health **at most 5**.
- There is exactly 1 exit tile in the grid.
- The agent moves one tile per step, except under the speed (Boots) power‑up, which doubles movement to two tiles per step.
- Each entity type has a fixed, known cost, damage, or reward (cost = –reward) where applicable:
  - Coins: reward 5.
  - Turn cost: 3 (Every turn incurs a cost of 3 (or reward -3) except for the final turn to win the game).
  - Lava: damage 2.
- Any key can be used to open any door, but each key can only be used once. 
- It is guaranteed that the initial grid will have an **equal number of keys and locked doors**
- It is also guaranteed that any doors would be **locked in the initial state.**
- There are **at most 5** gems in any grid.
- There are **at most 5** key-door pairs in any grid.
- There are **at most 5** pushable boxes in any grid.
- There can be **any number of** coins or lava.
- Power‑up durations are fixed:
  - Boots (Speed): 5 steps.
  - Ghost (Phasing): 5 steps.
  - Shield: 5 uses.
- There are **at most 3** of each powerup type in the grid.
- There will be **no overlapping entities in the initial state** of the grid (i.e each cell contains only `FloorEntity`, or `FloorEntity` with exactly 1 other entity)
- The maximum grid size is in the graded test cases will be **15x15**

````python
# Input states for Agent step function
from grid_adventure.grid import GridState
from grid_adventure.env import ImageObservation
from grid_adventure.entities import *
# State steppers
from grid_adventure.step import Action
from grid_adventure.grid import step as grid_step
# Utility helpers
import random
import os
from PIL import Image as PILImage
from typing import NamedTuple
import heapq
import numpy as np

_CLASSES = ("floor", "wall", "lava", "coin", "gem", "ghost", "human",
            "key", "exit", "locked", "box", "boots", "shield")

_FEAT_MU = np.array([+86.1971, +93.7115, +5.7803, +0.4007, +0.2004, +0.5753, +0.2334, +0.5317, +1.9727, +2.0344, +0.6919, +0.6521, +0.6258, +0.3648, +4.4249, +0.0769, +0.0661, +0.0594, +0.1151, +0.0519, +0.0907, +0.0519, +0.0622, +0.0626, +0.0282])
_FEAT_SD = np.array([+22.8624, +26.8194, +1.5991, +0.1046, +0.0772, +0.0664, +0.0389, +0.0509, +0.3248, +0.4545, +0.0944, +0.0796, +0.0844, +0.0537, +3.3137, +0.0273, +0.0206, +0.0287, +0.0461, +0.0202, +0.0301, +0.0177, +0.0173, +0.0205, +0.0159])
_LOGREG_W = np.array([
    [-1.4389, +0.2566, +0.0574, -0.5726, -0.3090, +0.5909, +0.2350, +0.2003, +0.4673, -0.0897, +0.2275, +0.2144, +0.1608],
    [-1.0079, +0.2631, -0.0527, -0.6191, +0.6122, +0.9940, +0.0195, +0.0481, +0.4655, -0.1889, -0.0895, +0.0210, -0.4653],
    [+0.3107, -0.2201, -0.5622, +0.1880, +0.9049, +0.0708, +0.6011, -0.4973, -0.2851, +0.4178, -0.4538, -0.5778, +0.1031],
    [+0.1614, +0.3574, +0.1706, +0.1509, +0.4256, -0.2452, -0.6511, +0.3168, -0.3921, -0.0974, +0.0775, +0.4451, -0.7195],
    [+0.2107, +0.3538, +0.6856, -0.3462, +0.1660, -0.3824, -0.2694, +0.2947, -0.0630, +0.0513, +0.4204, -0.4748, -0.6465],
    [+0.0045, -0.1614, -0.7016, +1.2107, +0.4062, -0.3938, -0.8824, -0.6865, -0.2860, -0.0429, -0.1194, +0.4890, +1.1637],
    [+0.1880, +0.0739, +0.1210, +0.0617, +0.0799, -0.4518, -0.0273, +0.3930, -0.0253, +0.4504, -0.0692, +0.3256, -1.1201],
    [-0.0068, +0.1013, -0.2690, +0.0493, +0.4270, +0.2258, -0.2476, +0.5083, +0.0468, -1.4659, +0.2456, -0.2711, +0.6562],
    [-0.0748, +0.9340, -0.4392, -0.1409, -0.2278, -0.1233, -0.4045, -0.4669, -0.0072, +0.1890, +0.4435, -0.0247, +0.3427],
    [-0.3527, -0.5085, -0.2146, -0.2649, -0.4122, +0.3604, -0.6473, +1.1165, -0.2998, +0.3128, +0.3914, -0.0234, +0.5424],
    [-0.3375, +0.4950, -1.1031, -0.1374, +0.2764, -0.9679, -0.1146, +0.2853, +0.4415, +1.4294, +0.8311, -1.1648, +0.0665],
    [-0.1105, -0.7051, -0.5456, -0.2451, +0.3407, -0.5761, -0.0381, -0.3116, +0.5691, -0.3757, +1.1486, +0.3710, +0.4782],
    [-0.0935, -0.4152, -0.0871, +0.8563, +0.1492, -0.1414, +0.1754, -0.6825, -0.1214, -0.0094, +0.3827, -0.4296, +0.4166],
    [-0.4518, +0.5839, +0.0729, -0.0651, -1.3520, -0.2915, +0.5164, +0.0342, -0.0467, +0.1000, +0.2816, +0.0578, +0.5603],
    [-0.0448, -0.0515, -0.1102, +0.0691, -1.3629, +0.7420, +0.3583, +0.0496, +0.1626, -0.1637, -0.0702, +0.0249, +0.3968],
    [-0.3434, -0.2034, -0.1455, +0.1605, +0.7987, -0.4964, -0.6752, +0.6118, -0.6341, +0.1697, +0.0264, +0.7302, +0.0009],
    [-0.1824, -0.4459, -0.0625, -0.2039, +0.3086, +0.0569, -0.6103, +0.2308, -0.6564, +0.5927, +0.1863, +0.7960, -0.0099],
    [-0.0140, +0.0601, +0.3546, +1.3789, -0.0494, -0.6360, -0.4165, -0.6426, +0.2596, -0.1370, +0.1897, -0.5318, +0.1844],
    [-0.1846, +0.0274, -0.2403, -0.5838, +0.9689, +0.2953, +0.6355, +0.1686, +0.1651, -0.5330, -0.1403, -0.2528, -0.3260],
    [+0.0083, +0.0462, +0.0923, +0.6134, -0.0073, +0.3800, -0.5349, -0.7420, +0.1368, +0.4789, -0.0393, -0.0340, -0.3986],
    [-0.0004, -0.3214, -0.0401, -0.3255, +0.7240, -0.2597, +0.1325, +0.5931, +0.3575, +0.6857, -0.4384, -0.4893, -0.6180],
    [+0.1323, +0.1262, +0.1181, +0.6141, +0.3694, +0.2406, -0.6464, -0.3560, -0.0692, -0.4818, +0.2431, +0.1539, -0.4443],
    [+0.0271, +0.0470, -0.4151, +0.3172, -0.2937, -0.3671, +0.9607, -0.1881, -0.7476, +0.2825, -0.1821, +0.4931, +0.0661],
    [-0.0086, +0.0004, -0.1626, +0.2896, -0.5960, -0.5307, +0.8504, +0.3993, -0.5362, -0.0051, +0.0083, +0.7698, -0.4788],
    [+0.3242, +0.6474, +0.4491, -0.2873, -0.8172, -0.1975, -0.6926, -0.2794, +0.0282, +0.3448, +0.3793, +0.5600, -0.4591],
    [-0.4938, -0.4390, -0.3533, -0.5375, +1.1729, -0.1311, +0.4070, +0.0869, -0.1525, +0.4637, -0.4231, +0.2708, +0.1291]
])

class MazeInfo(NamedTuple):
    human_pos: tuple
    health: int
    gems: frozenset
    keys: frozenset
    collected_keys: int
    doors: frozenset
    boxes: frozenset
    coins: frozenset
    active_buffs: tuple
    speed: frozenset
    shield: frozenset
    phasing: frozenset


class Agent:
    #-------- Helper Functions --------
    MOVES = {
        Action.UP: (0, -1),
        Action.DOWN: (0, 1),
        Action.LEFT: (-1, 0),
        Action.RIGHT: (1, 0),
    }

    def manhattan(self, a, b):
        return abs(a[0] - b[0]) + abs(a[1] - b[1])

    #-------- Extracting states --------
    def extract_state_info(self, state: GridState):
        human_pos = None
        health = 0
        gemset = set(); keyset = set(); doorset = set(); coinset = set(); boxset = set()
        speedset = set(); shieldset = set(); phasingset = set()
        collected_keysint = 0
        self.grid_width = state.width
        self.grid_height = state.height
        for x in range(state.width):
            for y in range(state.height):
                entities = state.objects_at((x, y))
                for e in entities:
                    if isinstance(e, AgentEntity):
                        human_pos = (x, y)
                        health = e.health.current_health
                        collected_keysint = sum(1 for item in e.inventory_list if isinstance(item, KeyEntity))
                    elif isinstance(e, GemEntity):            gemset.add((x, y))
                    elif isinstance(e, ExitEntity):           self.exit_pos = (x, y)
                    elif isinstance(e, KeyEntity):            keyset.add((x, y))
                    elif isinstance(e, LockedDoorEntity):     doorset.add((x, y))
                    elif isinstance(e, CoinEntity):           coinset.add((x, y))
                    elif isinstance(e, BoxEntity):            boxset.add((x, y))
                    elif isinstance(e, WallEntity):           self.walls.add((x, y))
                    elif isinstance(e, LavaEntity):           self.lava.add((x, y))
                    elif isinstance(e, SpeedPowerUpEntity):   speedset.add((x, y))
                    elif isinstance(e, PhasingPowerUpEntity): phasingset.add((x, y))
                    elif isinstance(e, ShieldPowerUpEntity):  shieldset.add((x, y))
        return MazeInfo(
            human_pos=human_pos, health=health,
            gems=frozenset(gemset), keys=frozenset(keyset),
            collected_keys=collected_keysint,
            doors=frozenset(doorset), boxes=frozenset(boxset), coins=frozenset(coinset),
            active_buffs=(0, 0, 0),
            speed=frozenset(speedset), shield=frozenset(shieldset), phasing=frozenset(phasingset),
        )

    #-------- Image Processing --------
    def shapify(self, gray):
        gx = np.zeros_like(gray)
        gx[1:-1, 1:-1] = (
            -gray[:-2, :-2] - 2 * gray[:-2, 1:-1] - gray[:-2, 2:]
            + gray[2:, :-2] + 2 * gray[2:, 1:-1] + gray[2:, 2:]
        )
        gy = np.zeros_like(gray)
        gy[1:-1, 1:-1] = (
            -gray[:-2, :-2] - 2 * gray[1:-1, :-2] - gray[2:, :-2]
            + gray[:-2, 2:] + 2 * gray[1:-1, 2:] + gray[2:, 2:]
        )
        return np.sqrt(gx ** 2 + gy ** 2)

    def _tile_features(self, edges):
        """25 selected features of a 32x32 Sobel edge-magnitude map."""
        e = edges.astype(np.float64)
        h, w = e.shape
        total = e.sum() + 1e-8
        mean = e.mean()
        std = e.std()
        peak_ratio = e.max() / (mean + 1e-8)

        border_frac = (total - e[6:26, 6:26].sum()) / total
        ring_border_frac = (total - e[4:28, 4:28].sum()) / total
        ring_mid_frac = (e[4:28, 4:28].sum() - e[10:22, 10:22].sum()) / total

        q_br = e[16:, 16:].sum() / total
        top_frac = e[:16, :].sum() / total

        row_sums = e.sum(1)
        col_sums = e.sum(0)
        row_peak = row_sums.max() / (row_sums.mean() + 1e-8)
        col_peak = col_sums.max() / (col_sums.mean() + 1e-8)

        sym_lr = 1.0 - (np.abs(e - e[:, ::-1]).sum() / (2 * total))
        sym_ud = 1.0 - (np.abs(e - e[::-1, :]).sum() / (2 * total))
        sym_diag = 1.0 - (np.abs(e - e.T).sum() / (2 * total))

        coverage = (e > mean).sum() / e.size

        center_small = e[12:20, 12:20].mean()
        border_small = (e[:4, :].sum() + e[-4:, :].sum()
                        + e[:, :4].sum() + e[:, -4:].sum()) / \
                       (2 * 4 * h + 2 * 4 * (w - 8))
        center_to_border = center_small / (border_small + 1e-8)

        # 4x4 block grid (each 8x8), row-major
        blocks = e.reshape(4, 8, 4, 8).mean(axis=(1, 3))
        block_sum = blocks.sum() + 1e-8
        b1  = blocks[0, 1] / block_sum
        b2  = blocks[0, 2] / block_sum
        b4  = blocks[1, 0] / block_sum
        b5  = blocks[1, 1] / block_sum
        b8  = blocks[2, 0] / block_sum
        b10 = blocks[2, 2] / block_sum
        b11 = blocks[2, 3] / block_sum
        b13 = blocks[3, 1] / block_sum
        b14 = blocks[3, 2] / block_sum
        b15 = blocks[3, 3] / block_sum

        return np.array([mean, std, peak_ratio,
                         border_frac, ring_border_frac, ring_mid_frac,
                         q_br, top_frac,
                         row_peak, col_peak,
                         sym_lr, sym_ud, sym_diag,
                         coverage, center_to_border,
                         b1, b2, b4, b5, b8, b10, b11, b13, b14, b15])

    def classify_tile(self, tile_rgb):
        target_size = (32, 32)
        tile_pil = PILImage.fromarray(tile_rgb).resize(target_size, PILImage.LANCZOS)
        tile_arr = np.array(tile_pil, dtype=float)
        tile_gray = tile_arr.mean(axis=2)
        tile_edges = self.shapify(tile_gray)

        f = self._tile_features(tile_edges)
        f_norm = (f - _FEAT_MU) / _FEAT_SD
        f_with_bias = np.concatenate([f_norm, [1.0]])
        scores = f_with_bias @ _LOGREG_W
        return _CLASSES[int(np.argmax(scores))]

    #-------- Planning --------
    def parse(self, observation) -> GridState:
        from grid_adventure.movements import CardinalMovement
        from grid_adventure.objectives import CollectGemsAndExitObjective

        info = observation["info"]
        image = observation["image"]

        grid_w = info["config"]["width"]
        grid_h = info["config"]["height"]
        img_h, img_w = image.shape[:2]
        tile_w = img_w / grid_w
        tile_h = img_h / grid_h
        img_rgb = image[:, :, :3]

        grid_data = {}
        for gx in range(grid_w):
            for gy in range(grid_h):
                x1 = int(round(gx * tile_w)); x2 = int(round((gx + 1) * tile_w))
                y1 = int(round(gy * tile_h)); y2 = int(round((gy + 1) * tile_h))
                grid_data[(gx, gy)] = self.classify_tile(img_rgb[y1:y2, x1:x2])

        gs = GridState(
            width=grid_w, height=grid_h,
            movement=CardinalMovement(),
            objective=CollectGemsAndExitObjective(),
            seed=info["config"]["seed"],
            turn_limit=info["config"]["turn_limit"],
        )

        for pos, etype in grid_data.items():
            if etype == "wall":      gs.add(pos, WallEntity())
            elif etype == "lava":    gs.add(pos, LavaEntity())
            elif etype == "coin":    gs.add(pos, CoinEntity())
            elif etype == "gem":     gs.add(pos, GemEntity())
            elif etype == "ghost":   gs.add(pos, PhasingPowerUpEntity())
            elif etype == "human":
                agent = AgentEntity()
                agent.set_health(info["agent"]["health"]["current_health"])
                for item in info["agent"]["inventory"]:
                    if isinstance(item, dict) and item.get("type") == "key":
                        agent.add_to_inventory(KeyEntity())
                    elif isinstance(item, str) and item == "key":
                        agent.add_to_inventory(KeyEntity())
                gs.add(pos, agent)
            elif etype == "key":     gs.add(pos, KeyEntity())
            elif etype == "exit":    gs.add(pos, ExitEntity())
            elif etype == "locked":  gs.add(pos, LockedDoorEntity())
            elif etype == "box":     gs.add(pos, BoxEntity())
            elif etype == "boots":   gs.add(pos, SpeedPowerUpEntity())
            elif etype == "shield":  gs.add(pos, ShieldPowerUpEntity())
        return gs

    def check_constraints(self, current_pos, target_pos, state: MazeInfo):
        speed_left, shield_left, phase_left = state.active_buffs
        health = state.health
        isPhasing = phase_left > 0
        hasShield = shield_left > 0
        x, y = target_pos
        if not (0 <= x < self.grid_width and 0 <= y < self.grid_height):
            return False
        if target_pos in self.walls and not isPhasing:
            return False
        if target_pos in state.doors and not isPhasing:
            return False
        if target_pos in self.lava:
            if not isPhasing and not hasShield and health - 2 <= 0:
                return False
        if target_pos in state.boxes and not isPhasing:
            dx = target_pos[0] - current_pos[0]
            dy = target_pos[1] - current_pos[1]
            box_end_pos = (target_pos[0] + dx, target_pos[1] + dy)
            if (box_end_pos in self.walls
                    or box_end_pos in state.boxes
                    or box_end_pos in state.doors):
                return False
        return True

    def state_after_turn(self, state: MazeInfo, action: Action, target_door=None):
        turn_cost = 3
        new_human_pos = state.human_pos
        new_health = state.health
        new_gems = set(state.gems); new_keys = set(state.keys)
        new_collected_keys = state.collected_keys
        new_doors = set(state.doors); new_boxes = set(state.boxes); new_coins = set(state.coins)
        speed_left, shield_left, phase_left = state.active_buffs
        new_speed = set(state.speed); new_shield = set(state.shield); new_phasing = set(state.phasing)

        if action == Action.PICK_UP:
            if new_human_pos in new_gems:   new_gems.remove(new_human_pos)
            if new_human_pos in new_keys:
                new_keys.remove(new_human_pos); new_collected_keys += 1
            if new_human_pos in new_coins:
                new_coins.remove(new_human_pos); turn_cost -= 5
            if new_human_pos in new_speed:
                new_speed.remove(new_human_pos); speed_left = 5
            if new_human_pos in new_shield:
                new_shield.remove(new_human_pos); shield_left = 5
            if new_human_pos in new_phasing:
                new_phasing.remove(new_human_pos); phase_left = 5
        elif action == Action.USE_KEY:
            door_to_remove = None
            for move in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                adj = (new_human_pos[0] + move[0], new_human_pos[1] + move[1])
                if adj in new_doors:
                    door_to_remove = adj; break
            if door_to_remove and new_collected_keys > 0:
                new_doors.remove(door_to_remove)
                new_collected_keys -= 1
                turn_cost = 3
            else:
                turn_cost = 500
        elif action in self.MOVES:
            iteration = 2 if speed_left > 0 else 1
            for _ in range(iteration):
                dx, dy = self.MOVES[action]
                step_target = (new_human_pos[0] + dx, new_human_pos[1] + dy)
                temp_state = state._replace(
                    human_pos=new_human_pos, health=new_health,
                    collected_keys=new_collected_keys,
                    doors=frozenset(new_doors), boxes=frozenset(new_boxes),
                    active_buffs=(speed_left, shield_left, phase_left),
                )
                if not self.check_constraints(new_human_pos, step_target, temp_state):
                    break
                new_human_pos = step_target
                if (new_human_pos in new_boxes) and (phase_left <= 0):
                    box_end_pos = (new_human_pos[0] + dx, new_human_pos[1] + dy)
                    new_boxes.remove(new_human_pos); new_boxes.add(box_end_pos)
                if new_human_pos in self.lava and phase_left <= 0:
                    if shield_left > 0:
                        shield_left -= 1
                    else:
                        new_health -= 2
                if new_human_pos == self.exit_pos:
                    turn_cost = 0
        speed_left = max(0, speed_left - 1)
        phase_left = max(0, phase_left - 1)
        return MazeInfo(
            human_pos=new_human_pos, health=new_health,
            gems=frozenset(new_gems), keys=frozenset(new_keys),
            collected_keys=new_collected_keys,
            doors=frozenset(new_doors), boxes=frozenset(new_boxes), coins=frozenset(new_coins),
            active_buffs=(speed_left, shield_left, phase_left),
            speed=frozenset(new_speed), shield=frozenset(new_shield), phasing=frozenset(new_phasing),
        ), turn_cost

    #-------- Agent --------
    def __init__(self):
        self.plan = []
        self.walls = set()
        self.lava = set()
        self.exit_pos = None
        self.visited = {}

    def step(self, state: GridState | ImageObservation) -> Action:
        if isinstance(state, GridState):
            if not self.plan:
                self.walls = set(); self.lava = set(); self.exit_pos = None
                self.plan = self.ComputeActions(state)
            if self.plan:
                return self.plan.pop(0)
            return Action.WAIT
        if not self.plan:
            gridstate = self.parse(state)
            self.walls = set(); self.lava = set(); self.exit_pos = None
            self.plan = self.ComputeActions(gridstate)
        if self.plan:
            return self.plan.pop(0)
        return Action.WAIT

    def microAstar(self, start_state: MazeInfo, goal_pos: tuple):
        speed_mult = 1.5 if start_state.active_buffs[0] > 0 else 3
        start_h = self.manhattan(start_state.human_pos, goal_pos) * speed_mult
        pq = [(start_h, 0, start_state, [])]
        visited = {}
        while pq:
            f, g, curr_state, actions = heapq.heappop(pq)
            if curr_state.human_pos == goal_pos:
                return actions, g, curr_state
            state_key = (
                curr_state.human_pos, curr_state.active_buffs, curr_state.collected_keys,
                curr_state.boxes, curr_state.coins, curr_state.health,
                curr_state.speed, curr_state.shield, curr_state.phasing,
                curr_state.gems, curr_state.keys,
            )
            if state_key in visited and visited[state_key] <= g:
                continue
            visited[state_key] = g
            pos = curr_state.human_pos
            has_collectible = (pos in curr_state.coins or pos in curr_state.speed
                               or pos in curr_state.shield or pos in curr_state.phasing
                               or pos in curr_state.keys or pos in curr_state.gems)
            if has_collectible:
                pickup_state, pickup_cost = self.state_after_turn(curr_state, Action.PICK_UP)
                new_g = g + pickup_cost
                sm = 1.5 if pickup_state.active_buffs[0] > 0 else 3
                new_f = new_g + self.manhattan(pickup_state.human_pos, goal_pos) * sm
                heapq.heappush(pq, (new_f, new_g, pickup_state, actions + [Action.PICK_UP]))
            for move in self.MOVES.keys():
                next_state, cost = self.state_after_turn(curr_state, move)
                if next_state.health <= 0 or next_state.human_pos == curr_state.human_pos:
                    continue
                new_g = g + cost
                sm = 1.5 if next_state.active_buffs[0] > 0 else 3
                new_f = new_g + self.manhattan(next_state.human_pos, goal_pos) * sm
                heapq.heappush(pq, (new_f, new_g, next_state, actions + [move]))
        return None, float('inf'), start_state

    def macroAstar(self, initial_state: MazeInfo):
        pq = [(0, 0, 0, initial_state, [])]
        macro_visited = {}
        state_id = 0

        while pq:
            f, g, _, state, plan = heapq.heappop(pq)

            if not state.gems and state.human_pos == self.exit_pos:
                return plan

            core = (
                state.human_pos,
                state.health,
                state.gems,
                state.keys,
                state.collected_keys,
                state.doors,
                state.boxes,
                state.coins,
                state.active_buffs,
                state.speed,
                state.shield,
                state.phasing,
            )
            if core in macro_visited and macro_visited[core] <= g:
                continue
            macro_visited[core] = g

            targets = []

            for key_pos in sorted(state.keys, key=lambda k: self.manhattan(state.human_pos, k)):
                targets.append(("KEY", key_pos))

            for gem_pos in sorted(state.gems, key=lambda gm: self.manhattan(state.human_pos, gm)):
                targets.append(("GEM", gem_pos))

            if state.collected_keys > 0:
                for door_pos in state.doors:
                    for dx, dy in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                        adj = (door_pos[0] + dx, door_pos[1] + dy)
                        if (
                            0 <= adj[0] < self.grid_width
                            and 0 <= adj[1] < self.grid_height
                            and adj not in self.walls
                            and adj not in state.doors
                            and adj not in state.boxes
                        ):
                            targets.append(("DOOR", adj, door_pos))

            for coin_pos in sorted(state.coins, key=lambda c: self.manhattan(state.human_pos, c)):
                targets.append(("COIN", coin_pos))

            for p_type, p_set in [("BOOTS", state.speed), ("SHIELD", state.shield), ("GHOST", state.phasing)]:
                for p_pos in sorted(p_set, key=lambda p: self.manhattan(state.human_pos, p)):
                    targets.append((p_type, p_pos))

            if not state.gems:
                targets.append(("EXIT", self.exit_pos))

            for t in targets:
                t_type = t[0]
                t_pos = t[1]
                door_actual_pos = t[2] if t_type == "DOOR" else None

                move_actions, move_cost, arrived_state = self.microAstar(state, t_pos)
                if move_actions is None:
                    continue

                if t_type == "DOOR":
                    if arrived_state.collected_keys > 0 and door_actual_pos in arrived_state.doors:
                        final_state, int_cost = self.state_after_turn(arrived_state, Action.USE_KEY)
                        final_actions = move_actions + [Action.USE_KEY]
                        if door_actual_pos in final_state.doors:
                            continue
                    else:
                        continue

                elif t_type in ("GEM", "KEY", "BOOTS", "SHIELD", "GHOST", "COIN"):
                    already_collected = False
                    if t_type == "COIN" and t_pos not in arrived_state.coins:
                        already_collected = True
                    elif t_type == "GEM" and t_pos not in arrived_state.gems:
                        already_collected = True
                    elif t_type == "KEY" and t_pos not in arrived_state.keys:
                        already_collected = True
                    elif t_type == "BOOTS" and t_pos not in arrived_state.speed:
                        already_collected = True
                    elif t_type == "SHIELD" and t_pos not in arrived_state.shield:
                        already_collected = True
                    elif t_type == "GHOST" and t_pos not in arrived_state.phasing:
                        already_collected = True

                    if already_collected:
                        final_state = arrived_state
                        int_cost = 0
                        final_actions = move_actions
                    else:
                        final_state, int_cost = self.state_after_turn(arrived_state, Action.PICK_UP)
                        final_actions = move_actions + [Action.PICK_UP]

                else:
                    final_state = arrived_state
                    int_cost = 0
                    final_actions = move_actions

                new_g = g + move_cost + int_cost

                h = 0
                if final_state.gems:
                    nearest_gem_dist = min(self.manhattan(final_state.human_pos, gm) for gm in final_state.gems)
                    farthest_gem_to_exit = max(self.manhattan(gm, self.exit_pos) for gm in final_state.gems)
                    h += nearest_gem_dist * 3
                    h += len(final_state.gems) * 3
                    h += farthest_gem_to_exit * 3
                else:
                    h += self.manhattan(final_state.human_pos, self.exit_pos) * 3

                if t_type == "COIN":
                    h += 2
                elif t_type in ("BOOTS", "SHIELD", "GHOST"):
                    h += 1

                state_id += 1
                heapq.heappush(pq, (new_g + h, new_g, state_id, final_state, plan + final_actions))

        return []

    def ComputeActions(self, state: GridState):
        self.walls = set(); self.lava = set()
        maze_info = self.extract_state_info(state)
        if maze_info.human_pos is None or self.exit_pos is None:
            return [Action.WAIT]
        final_plan = self.macroAstar(maze_info)
        if not final_plan:
            return [Action.WAIT]
        return final_plan

    def info(self) -> dict[str, str]:
        return {"name": "Random AI Agent"}
````

### Hardcoding Your Trained ML Model
Because Coursemology disallows arbitrary auxiliary file uploads (e.g. separate `.pth` / `.pkl` weight files), any ML models your `Agent` depends on must ship **inline inside your submission**. To make this easy (and less error‑prone than manually copying numbers), we provide two helper functions in `utils.py`:

- `generate_torch_loader_snippet(model, example_inputs=..., prefer='auto', compression='zlib')`
- `generate_sklearn_loader_snippet(model, compression='zlib')`

Each returns a fully self‑contained Python source string defining a `get_model(...)` function. You copy‑paste that function into your agent code. At runtime the function reconstructs the trained model from a Base64‑encoded (optionally compressed) blob embedded directly in the source.

Below we show the end‑to‑end workflow for PyTorch and scikit‑learn. We also keep (optional) manual hardcoding examples for learning.

Suppose you train a model on the following data:

````python
import numpy as np

X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.array([1, 8, 2, 4])
````

#### PyTorch

Assume you have the following simple network (we'll train it briefly, then auto‑generate a loader snippet):

````python
import torch
from torch import nn

net = nn.Linear(2, 1)
````

Train the network:

````python
def train(net: nn.Module, X: np.ndarray, y: np.ndarray):
    X = torch.tensor(X, dtype=torch.float32)
    y = torch.tensor(y[:, None], dtype=torch.float32)
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(net.parameters(), lr=0.01)
    for epoch in range(100):
        optimizer.zero_grad()
        output = net(X)
        loss = criterion(output, y)
        loss.backward()
        optimizer.step()

train(net, X, y)
````

**Option 1 (Recommended):** Automatically generate a self‑contained loader snippet.

Run the following after training:

````python
# Auto‑generate a loader snippet for the trained PyTorch model
from utils import generate_torch_loader_snippet

example_input = torch.tensor([[0.0, 0.0]], dtype=torch.float32)  # minimal example for tracing if needed
snippet = generate_torch_loader_snippet(net, example_inputs=example_input, prefer="auto", compression="zlib")
print(snippet)

# You would copy the FULL printed function definition (def get_model(...): ...) into your agent source code.
# Then inside Agent.__init__ you could do: self.model = get_model()
````

**Option 2 (Manual, Educational):** Manually copy weights into a freshly constructed model. This is more error‑prone and not recommended for larger networks, but shown for completeness.


````python
# Set precision of printing numbers to 10 d.p.
torch.set_printoptions(precision=10)
print(net.state_dict())
````

Example output:
```python
OrderedDict([('weight', tensor([[0.1491376609, 1.3386186361]])), ('bias', tensor([0.6359716654]))])
```

Notes:
- `OrderedDict` is from the `collections` module; `tensor` is a PyTorch tensor constructor.
- Ensure you exactly match shapes and dtypes.

To hardcode manually:

````python
from collections import OrderedDict
from torch import tensor

# declare the same model
model = nn.Linear(2, 1)

# Hardcode the weights from the previous step (input your values here)
coeffs = OrderedDict([('weight', tensor([[0.1491376609, 1.3386186361]])), ('bias', tensor([0.6359716654]))])

# Load the model with the hardcoded weights
model.load_state_dict(coeffs)

# The model now has the same weights as it was previously trained
X_tensor = torch.tensor(X, dtype=torch.float32)

# Check if the outputs are the same
torch.allclose(model(X_tensor), net(X_tensor))
````
